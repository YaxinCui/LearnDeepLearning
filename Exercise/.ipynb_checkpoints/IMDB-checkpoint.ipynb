```python
# 读取glove词向量，
def read_glove(glove_path):
    word2index={}
    wordEmbed = []
    with open(glove_path, 'r', encoding='utf8') as f:
        for line in progressbar(f.readlines()):
            line = line.split()
            word2index[line[0]] = len(word2index)
            wordEmbed.append(line[1:])
    wordEmbed = np.asarray(wordEmbed, np.float32)
    return wordEmbed, word2index
wordEmbed, word2index = read_glove('../glove/glove.6B/glove.6B.100d.txt')
```

```python
# 读取数据
label2index = {'neg':0, 'pos':1}

class IMDBDatasset(Dataset):
    def __init__(self, file_path):
        self.file_path = file_path
        self.negPath = os.path.join(file_path, 'neg')
        self.posPath = os.path.join(file_path, 'pos')
        
        self.negDirList = os.listdir(self.negPath)
        self.posDirList = os.listdir(self.posPath)

        self.lenNeg = len(self.negDirList)
        self.lenPos = len(self.posDirList)
        
        self.length = self.lenNeg + self.lenPos
        
        self.lengthList = None
        
        print("共有pos数据 {} 条, neg数据 {} 条，共 {} 条".format(self.lenPos, self.lenNeg, self.length))
        
    def __getitem__(self, index):
        if index < self.lenNeg:
            with open(os.path.join(self.negPath, self.negDirList[index]), 'rb') as f:
                sentence = f.read()
            return (sentence, label2index['neg'])
        
        else:
            with open(os.path.join(self.posPath, self.posDirList[index-self.lenNeg]), 'rb') as f:
                sentence = f.read()
            return (sentence, label2index['pos'])
        
    def __len__(self):
        return self.length
    
    def show_data_distribution(self, space_list=[2*i for i in range(128)]):
        # 展示数据分布
        print("分析数据单词长度")
        if self.lengthList is None:
            self.lengthList = []
            for i in progressbar(range(self.length)):
                self.lengthList.append(len((self.__getitem__(i)[0]).split()))
            plt.hist(self.lengthList, space_list)
        else:
            plt.hist(self.lengthList, space_list)
        
    def print_sample(self):
        index = randint(0, self.length - 1)
        print(self.__getitem__(index))
```

```python
trainIMDBDataset = IMDBDatasset('../IMDB/aclImdb_v1/aclImdb/train')
trainIMDBDataLoader = DataLoader(trainIMDBDataset, batch_size=8, shuffle=True)

testIMDBDataset = IMDBDatasset('../IMDB/aclImdb_v1/aclImdb/test')
testIMDBDataLoader = DataLoader(testIMDBDataset, batch_size=8, shuffle=True)
```

```python
dataloaders = {'train':trainIMDBDataLoader, 'val':testIMDBDataLoader}
def train_model(model, criterion, optimizer, scheduler, dataloaders, skip = 100, num_epochs=25):
    # 
    import time
    import copy
    
    since = time.time()
    
    best_model_wts = copy.deepcopy(model.state_dict())
    best_acc = 0.0
    
    for epoch in range(num_epochs):
        print('Epoch {}/{}'.format(epoch, num_epochs - 1))
        print('-' * 10)
        
        # 每个epoch都有一个训练和验证的阶段
        for phase in ['train', 'val']:
            if phase == 'train':
                scheduler.step()
                model.train()
            else:
                model.eval()
                
            running_loss = 0.0
            running_corrects = 0
            
            # 迭代数据
            for i, (inputs, labels) in enumerate(dataloaders[phase]):
                inputs = inputs.to(device)
                labels = labels.to(device)
                
                # 零参数梯度
                optimizer.zero_grad()
                
                # 前向
                # track history if only in train
                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs.float())
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)
                    
                    # 后向仅在训练阶段进行优化
                    if phase == 'train':
                        loss.backward()
                        optimizer.step()
                        
                # 统计
                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)
                
                if i % skip == 0:
                    print('{} Loss: {:.4f} Acc: {:.4f}={}/{}'.format(phase, loss.item(), torch.sum(preds == labels.data).double()/inputs.size(0), torch.sum(preds == labels.data), inputs.size(0)))

            epoch_loss = running_loss / len(dataloaders[phase])
            epoch_acc = running_corrects.double() / len(dataloaders[phase])
            
            print('{} Loss: {:.4f} Acc: {:.4f}'.format(phase, epoch_loss, epoch_acc))
            
            # 深度复制
            if phase == 'val' and epoch_acc > best_acc:
                best_acc = epoch_acc
                best_model_wts = copy.deepcopy(model.state_dict())
        print()
        
    time_elapsed = time.time() - since
    print('Training compete in {:.0f}m {:.0f}s'.format(time_elapsed // 60, time_elapsed % 60))
    print('Best val Acc: {:4f}'.format(best_acc))
    
    # 加载最佳模型权重
    model.load_state_dict(best_model_wts)
    return model
```

```python
# LSTM

class IMDBLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, max_length, label_size):
        super(IMDBLSTM, self).__init__()
        
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.max_length = max_length
        self.label_size = label_size
        
        # input_size应该等于词向量 embedding的维度
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, bias=True, batch_first=False,
                 dropout=0.2, bidirectional=False)
        
        # 用Outputs来预测结果
        self.linear = nn.Linear(num_layers * hidden_size, label_size)
        
    def forward(self, inputs):
        # inputs (seq_len, batch, input_size)
        h0 = torch.zeros(2, 3, 20) # (num_layers, batch, output_size)
        c0 = torch.zeros(2, 3, 20) # (num_layers, batch, output_size)
        
        outputs, (hn, cn) = self.lstm(inputs)
        # output： 维度和输入数据类似，只不过最后的feature部分会有点不同，即 (seq_len, batch, num_directions * hidden_size),
        # output.view(seq_len, batch, num_directions, hidden_size)用作分离方向，输出维度
        # h_n：(num_layers * num_directions, batch, hidden_size)，
        # c_n ：(num_layers * num_directions, batch, hidden_size)，只会输出最后个time step的cell状态结果（如下图所示）。
        # print("outputs {} hn {} cn {}".format(outputs.size(), hn.size(), cn.size()))
        # outputs = outputs.transpose(0, 1)
        # outputs = outputs.reshape(-1, self.max_length * self.hidden_size)
        # outputs = F.softmax(F.relu(self.linear(outputs), dim=1))
        
        return F.softmax(F.relu(self.linear(hn.transpose(0, 1).reshape(-1, self.num_layers * self.hidden_size))), dim=1)

imdblstm = IMDBLSTM(input_size=50, hidden_size=100, num_layers=2, max_length=256, label_size=2)
```

```python
def deal_batch_sentences(embeding, word2index, batch_sentences, max_length, word_dim):
#    print("1 ", datetime.datetime.now())
    word_set = set(word2index.keys())
    
    embed_sentences = torch.zeros(len(batch_sentences), max_length, word_dim)
    for i, sentence in enumerate(batch_sentences):
        # print("1 ", datetime.datetime.now())
        for j, word in enumerate(sentence.split()):
            if j >= max_length:
                break
            # print("2 ", datetime.datetime.now())
            if word in word_set:
                # print("3 ", datetime.datetime.now())
                embed_sentences[i][j] = embedding[word2index[word]]
                # print("4 ", datetime.datetime.now())

            else:
                continue
    
    return embed_sentences
```
